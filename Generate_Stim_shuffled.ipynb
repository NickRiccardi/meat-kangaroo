#### Shuffle the order of the phrases 10 times with predifined random seeds

Considering that LLM's responses are not fully reproducible and the responses may be affected by an order effect (the meaningfulness judgements of different phrases may not be indepdendent), the order of the 2160 phrases was shuffled for 10 times. The random seeds for the ten iterations were predifined to generate reproducible results.

```python
df_stim = pd.read_csv(os.path.join(Dir_project,'codes','Stim_2160.csv'))
np.random.seed(123)

List_seed = np.random.randint(1,10000,9)
dfs_stim = []
for i,seed in enumerate(List_seed):
    df = df_stim.copy()
    np.random.seed(seed)
    df['shuffleID'] = i+1
    df['shuffleSeed'] = seed
    df['rand'] = np.random.rand(len(df))
    df = df.sort_values(['rand']).reset_index(drop=True)
    df[['Idx','promptID','idxInPrompt']] = df_stim[['Idx','promptID','idxInPrompt']]
    dfs_stim.append(df)
df_stim['shuffleID'] = 0
dfs_stim.append(df_stim)
dfs_stim = pd.concat(dfs_stim,ignore_index=True)

dfs_stim.to_csv(os.path.join(Dir_project,'codes','Stim_2160_shuffle10.csv'),index=False)
dfs_stim
```
